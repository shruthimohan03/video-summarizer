Extractive summarization by selecting the most similar sentences iteratively.

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def extractive_summary(text, top_n):

    # Break it to sentences (Sentence Tokenization)
    sentences = text.split('. ')

    # Embed using SBERT
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(sentences)

    # Compute similarity matrix
    similarity_matrix = cosine_similarity(embeddings)

    # Find the most central sentence (highest avg similarity)
    avg_sim = np.mean(similarity_matrix, axis=1)
    central_idx = np.argmax(avg_sim) # np.argmax gives the index of the max value

    selected_indices = [central_idx] # storing the selected sentences' indices
    #print(sentences[central_idx],max(avg_sim))
    while len(selected_indices) < top_n:
    #while len(selected_indices)!=len(sentences):
        # Find the most similar sentence to the selected ones
        sim_scores = np.sum(similarity_matrix[selected_indices], axis=0)

        sim_scores[selected_indices] = -1  # Avoid selecting already chosen sentences
        next_idx = np.argmax(sim_scores)
        selected_indices.append(next_idx)
        #print(sentences[next_idx],max(sim_scores)/len(selected_indices))
    # Return extracted sentences in original order
    selected_indices.sort()
    return [sentences[i] for i in selected_indices]

# Example usage
with open('/content/transcribed_text_nptel_video_1.txt', 'r') as file:
    text = file.read()
summary = extractive_summary(text, 10)
print("\n".join(summary))
```

```python
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline

# Load the pretrained punctuation model
model_name = "nickmuchi/deberta-v3-small-punctuation-multilingual"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Create a pipeline for punctuation restoration
punctuation_pipeline = pipeline("token-classification", model=model, tokenizer=tokenizer)

def restore_punctuation(text):
    words = text.split()
    punctuated_text = ""
    
    for word in words:
        prediction = punctuation_pipeline(word)
        punctuated_text += word

        if prediction:  # If a punctuation is predicted, add it
            for p in prediction:
                if p['entity'].startswith('PUNCT'):
                    punctuated_text += p['word']

        punctuated_text += " "

    return punctuated_text.strip()

# Example transcript (raw text)
transcript = "hello how are you i am fine thank you what about you"
punctuated_transcript = restore_punctuation(transcript)

print("Original Transcript:", transcript)
print("Punctuated Transcript:", punctuated_transcript)
```

```python
import re 
def extract_sentences_with_keywords(text, keywords):
    # Split the input text into sentences
    sentences = re.split(r'(?<=[.!?])\s+', text)
    
    # Find sentences containing any of the keywords
    matching_sentences = [
        sentence for sentence in sentences
        if any(keyword in sentence.lower() for keyword in keywords)
    ]
    
    return matching_sentences
```

```python
important_sentences = extract_sentences_with_keywords(transcription2, educational_keywords)

# Print the results
print("Sentences containing educational keywords:")
for sentence in important_sentences:
    print("- " + sentence.strip())
```
